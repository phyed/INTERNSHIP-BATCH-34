```python
# FIRST CREATE EMPTY LISTS 
job_title=[]
job_location=[]
company_name=[]
experience_required=[]

#scraping job titles------
for i in driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]'):
    job_title.append(i.text)
    
#scraping job location---

for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 locWdth"]'):
    job_location.append(i.text)
    

# scraping campany name
for i in driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]'):
    company_name.append(i.text)
    
# scraping experience

for i in driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft fs12 lh16 expwdth"]'):
    experience_required.append(i.text)
    
# DF PRESENTATION 
import pandas as pd

df=pd.DataFrame({'TITLE':job_title,'LOCATION':job_location,'COMPANY_NAME': company_name,'EXPERIENCE REQUIRED':experience_required})
df.head(10)

    
```

```python
# SCRAPING THE URLS OF JOB TITLE
url=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
for i in url[0:10]:
    print(i.get_attribute('href'))
```

```python
# SCRAPING DATA FROM MULTIPLE PAGES
job_titles_multiple=[]
start=0 #STARTING PAGE
end=2 # ENDING PAGE
for page in range(start,end):
    titles=driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
    
    for i in titles:
        job_titles_multiple.append(i.text)
        
    next_button=driver.find_element(By.XPATH,'//a[@class="fright fs14 btn-secondary br2"]')
    next_button.click()
    time.sleep(3) # PRG WAITS FOR 3 S
```
